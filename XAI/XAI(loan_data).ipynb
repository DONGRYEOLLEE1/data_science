# XAI

## XAI란?
- XAI는 인공지능 모델이 특정 결론을 내리기까지 어떤 근거로 의사 결정을 내렸는지를 알 수 있게 설명 가능성을 추가하는 기법. 주로 머신러닝이과 연관된 사람이 시스템을 신뢰하기 위해 사용한다. 요즘의 인공지능은 점점 더 복잡해지고, 학습하는 데이터의 양이 사람이 해석 불가능할만큼 많아졌기 때문에 XAI는 모델의 복잡성을 해소하기 위해 등장했다. XAI는 해석 가능한 인공지능, 또는 투명한 인공지능이라고 부른다.
- 딥러닝을 비롯한 현대 머신러닝 알고리즘은 불투명한 편이다. XAI 기법 중 피쳐 중요도 표시, 필터 시각화 기법, LRP를 이용한 히트맵 해석 등이 불투명한 인공지능의 의사 결정 과정을 설명적으로 개선한다. 다르파의 지침에 의하면 XAI는 세가지의 과정을 포함한다.
    - 기존 머신러닝 모델에 설명 가능한 기능 추가
    - 머신러닝 모델에 HCI(Human Computer Interation) 기능 추가
    - XAI를 통한 현재 상황의 개선
    
## 시각화와 XAI의 차이
- XAI의 많은 산출물이 시각화 기법에 의존하고 있지만, 머신러닝 모델의 과정을 시각화했다고해서 모두 XAI라고 말할 수는 없다. XAI의 **핵심은 해석 가능성**이다.

### Interpretability
- 해석가능성은 왜 해당 모델을 신뢰해야 하는지, 아니면 하지 말아야 하는지, 모델이 왜 특정 결정을 했는지에 관한 근거를 찾고, 어떤 결과가 예상되는지 판단하는 과정이다.
- XAI는 대리 분석(Surrogate Analysis), 부분 의존성 플롯(Partial Dependence Plots, PDPs), 유사도 분석(Similarity Measure), 피쳐 중요도(Feature Importance)등의 기법으로 데이터와 모델을 설명한다.

### XAI의 작동원리
- XAI의 핵심사항은 다음과 같다.
    1. 모델을 설명하는 대상은 누구인가?
    2. 설명은 얼마나 정확하고 얼마나 정확하고 정밀해야 하는가?
    3. 설명하는 대상이 전체적인 모델인가, 특정 의사결정인가?

## Decision Tree
- 의사결정트리는 질문을 던지고 답을 하는 과정을 연쇄적으로 반복해 집단을 분류하거나 예측하는 전통적인 머신러닝 방법.
- 의사결정트리를 시각화하면 그 자체적으로 해석력을 가지고 있지만, XAI를 결합하면 일반인들도 해석 가능하게 만들 수 있음.
- 의사결정트리는 해법을 찾아가는 과정을 도식적으로 표현할 수 있어 연구자가 분석 과정을 쉽게 이해하고 설명할 수 있다는 장점이 있음.
    - 의사결정트리를 시각화함으로써 모델이 내리는 결정을 단계별로 이해할 수 있음.

- 의사결정트리를 시각화하면 모델이 어떤 비중으로 특정 피쳐를 우선순위로 생각하는지, 그리고 단계마다 어떤 기준으로 당뇨병을 진단하는지 순서대로 따라가 볼 수 있다. 따라서 의사결정트리에서 피쳐는 모델의 성능과 결정에 대한 이해를 높이는데 매우 중요함.

### PDPs
- 부분 의존성 플롯은 피쳐의 수치 변환에 따라 모델에 기여하는 정도가 어떻게 달라지는지 확인 할 수 있는 XAI 기법.
- 부분 의존성 플롯은 특정 피쳐의 값에 따라서 알고리즘 해석 능력이 얼마나 증가하고 감소하는지 관찰할 수 있음.

- 장점
    - PDP의 계산은 **직관적**이다. 특정 특성 값의 부분 의존도 함수는 모든 데이터 포인트가 해당 특성 값을 가정하도록 하는 경우 평균 예측을 나타낸다.

    - 부분 의존도를 계산한 특성이 다른 특성과 상관없는 경우 PDP는 해당 특성이 평균적으로 예측에 영향을 미치는 방식을 완벽하게 나타낸다. 상관관계가 없는 경우 **해석이 명확하다**. PDP 는 j번째 특성이 변경될 때, 데이터 집합의 평균 예측이 어떻게 변하는지 보여준다.

    - 부분 의존도는 **구현하기 쉽다**.

    - 부분 의존도 그림에 대한 계산에는 **원인 해석**이 있다. 우리는 특성에 개입하고 예측의 변화를 측정한다. 이를 통해 특성과 예측 사이의 인과관계를 분석한다.

- 단점

    - 부분 의존성 함수에서 **현실적인 최대 특성 수는 2개**이다. 이건 PDP의 잘못이 아니라 2차원 표현의 잘못이며 또한 3차원 이상을 상상할 수 없는 한계이다.

    - 일부 PD 그림에는 **특성 분포**가 표시되지 않는다. 데이터가 거의 없는 영역을 지나치게 해석할 수 있으므로 분포를 생략하는 것은 오해를 유발할 수 있다. 이 문제는 깔개(X축의 데이터 포인트에 대한 표시) 또는 히스토그램으로 쉽게 해결할 수 있다.

    - 독립성에 대한 **가정**은 PDP의 가장 큰 문제다. 

    - **PDP에는 평균 한계 효과만 표시되므로 이질적인 효과가 숨겨질 수도 있다.**

## 대리 분석(Surrogate Analysis)
- 대리분석이란 엔지니어링에 사용되던 용어였다. 이름에서 알 수 있듯이 본래 기능을 흉내 내는 간단한 대체재를 만들어서 프로토타입이 동작하는지 판단하는 분석 방법이다. 이 방법은 결과를 내기까지 너무 비용이 많이 들거나 연산이 많아 정확한 결과를 내는데 시간이 많이 소요될 때 사용됐다.

![ㅇㅇ](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FVIbcI%2FbtrniTkjT5U%2FsHiikiUJczer1yaykZkPh1%2Fimg.png)

- 대리분석의 핵심은 분석해야 할 모델 $f$를 흉내내는 모델 $g$를 만들어 학습하는 방법이다. 이때 모델 $g$는 모델 $f$의 학습 방식과 같을 수도 있고, 다를 수도 있다.
- 모델 $f$가 SVM을 사용한 모델이라면, 모델 $g$는 트리나 선형회귀모델일 수 있다.
- 모델 $g$를 결정하는 요소
    - 모델 $f$보다 학습하기 쉬워야 한다.
    - 설명이 가능하다.
    - 모델 $f$를 유사하게 흉내낼 수 있다.
    
- 모델 $g$를 학습시키는 과정은 두 가지.
    1. 모델 $f$를 학습시킬 때 처럼 학습 데이터 전부를 학습하는 방법
    2. 데이터를 라벨 별로, 학습 데이터의 일부를 추려 학습하는 방법
    
- 위의 과정이 끝나면 모델 $g$는 모델 $f$를 어설프게 따라한 것이다. 모델 $g$는 설명 가능하기 때문에 모델 $f$가 어떻게 학습됐을지 간단하게 해석할 수 있다. 다만 모델$g$는 모델$f$의 비해 정확도가 떨어진다.
- 학습데이터를 사용해서 대리 분석 모델을 글로벌 대리 분석이라고 하며, 데이터 하나를 해석하는 과정을 로컬대리분석 이라고 한다.

### 글로벌 대리 분석
- 전체 학습 데이터를 사용해 블랙박스 함수 $f$를 따라 하는 유사 함수 $g$를 만들고 $g$를 해석 가능하도록 변조하는 기법

- 장점
    - 다양한 XAI 기법을 자유롭게 적용할 수 있으며, 블랙박스 모델을 이해하고 있지 않아도 Measure function을 가지고 모델 $f$가 어떻게 학습됐는지 설명할 수 있을 정도로 유연하다.
    - 직관적
    - 원래 모델을 유사하게 따라하는 대리 모형을 찾을 수 있으면 큰 힘을 들이지 않고 모델을 해석할 수 있다.

- 단점
    - 모델 $g$는 모델 $f$를 유사하게 설명하는 모델이라는 점
    - measure function 설명 가능성 판단 기준이 주관적이라는 점
    - 글로벌 대리 분석에 사용되는 데이터 x가 편향됐을 위험이 있다는 점


### 로컬 대리 분석

- 데이터 하나에 대해 블랙박스가 해석하는 과정을 분석하는 기법
- LIME(Local Interpretable Model-agnostic Explanations, 학습 기법과 관계없이 모델을 설명할 수 있는 로컬 설명 가능 모델) 이라는 이름으로 알려져 있음

- 장점
    - 데이터 하나에 대해 원래 모델이 분류한 결과를 해석 가능하게 알려줌
    - 모델 학습 방법과 관계없이 적용할 수 있음

- 단점
    - 모델이 잘못된 결과를 출력할 수도 있음
    - 비전문가일 경우, 현실 데이터를 해석한 결과물을 받고 해석에 어려움을 겪을 가능성이 높음
    - 데이터 하나에 대해 분석하기 때문에 전체 데이터양에 따라 분석 시간이 다름

## LIME

- 로컬 대리 분석으로 데이터 하나에 대한 해석과정을 보여주는 알고리즘
- 학습 기법과 관계없이 모델을 설명할 수 있는 로컬 설명 가능 모델
- LIME은 현재 데이터의 어떤 영역을 집중해서 분석했고, 어떤 영역을 분류 근거로 사용했는지 알려주는 XAI 기법, 모델 학습 방법과 관계없이 적용할 수 있는 XAI다.
- 텍스트 분석 , 이미지 분석

![ㅇㅇ](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FwmEqm%2FbtrnkvwhRzK%2FZWzoCVjDx4zkHwAgMyXDk1%2Fimg.png)

- 장점
    - 머신러닝 알고리즘에 관계없는 XAI를 적용 가능
    - 매트릭스로 표현 가능한 데이터(텍스트, 이미지)에 대해 작동하는 XAI 기법
    - 다른 XAI 기법과 비교했을 때 매우 가벼움

- 단점
    - 슈퍼 픽섹을 구하는 알고리즘을 구하는 모델 $g$의 결정 경계를 확정 짓는 방법이 비결정적임
        - 비결정적이기 때문에 입력 값이 같아도 출력 결과가 항상 일정하지 않다
    - 데이터 하나에 대해 설명을 하기에 모델 전체에 대한 일관성을 보전하지 못함
        - 전체 데이터 신뢰도를 떨어뜨릴 수 있음

```python
# split to train/test

X = loan_data.loc[:, 'gender' : 'loan_term']
y = loan_data.loc[:, 'loan_status']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 200)
```

```python
# Train model

model = XGBClassifier().fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

def calculate_accuracy(model, x_test, y_test):
    y_pred = model.predict(x_test)
    predictions = [round(value) for value in y_pred]
    
    # evaluation
    accuracy = accuracy_score(y_test, predictions)
    print('Accuracy: %.2f%%' % (accuracy * 100.0))
    return accuracy
```

```python
calculate_accuracy(model, x_test, y_test)
```

```python
# 파라미터 조정

model = XGBClassifier(booster = 'gbtree',
                     learning_rate = .04,
                     n_estimators = 50,
                     reg_alpha = .45,
                     reg_lambda = .6).fit(x_train, y_train)

print(model.evals_result)
calculate_accuracy(model, x_test, y_test)
```
